```python
import numpy as np
from sklearn import datasets
import matplotlib.pylab as pl
iris = datasets.load_iris()

```

```python
X = iris.data
y = iris.target
X[:2,:]
```

```python
np.unique(y)
```

```python
%pylab inline
pl.scatter(X[:,0], X[:,1], c=y)
pl.xlabel("sepal length")
pl.ylabel("sepal width")
```

```python
from sklearn import decomposition
pca = decomposition.PCA(n_components=2)
pca.fit(X)
T = pca.transform(X)
```

```python
# visualizing the results of our Princiapal Components Analysis performed on the IRIS datasets
pl.scatter(T[:,0], T[:,1], c= y)
```

```python
from sklearn import neighbors
knn = neighbors.KNeighborsClassifier(7)
# 3 we are working with the two dimensional 
knn.fit(T,y)
knn.predict([[0.1, 0.2]])
yPred = knn.predict(T)
(yPred == y).mean()
```

```python
from sklearn import metrics
print(metrics.recall_score(yPred, y, average=None))
```

```python
print(metrics.classification_report(yPred, y))
```

```python
print(metrics.confusion_matrix(yPred, y))
```
