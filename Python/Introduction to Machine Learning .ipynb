```python
import numpy as np
def rot(angle, R = np.zeros((3,3))):
        cx, cy, cz = np.cos(angle)
        sx, sy, sz = np.sin(angle)
        R.flat = (cx*cz - sx*cy*sz, cx*sz + sx*cy*cz, sx*sy, -sx*cz - cx*cy*sz, -sx*sz + cx*cy*cz, cx*sy, sy*sz, -sy*cz, cy)
        return R
```

```python
# plotting swiss rolls in 3 D
import mpl_toolkits.mplot3d.axes3d as p3
import pylab as pl
def myplot(X, col = "blue"):
    fig = pl.figure()
    ax = p3.Axes3D(fig)
    ax.view_init(7, -80)
    ax.scatter( X[:, 0], X[:, 1], X[:, 2], s= 10, c=col, marker="o")
```

```python
rotate_swiss_roll = rot((0,np.pi/4,np.pi/4))
```

```python
from sklearn.datasets import make_swiss_roll
swiss_roll1, _ = make_swiss_roll(1000)
swiss_roll2, _ = make_swiss_roll(1000)

X = np.concatenate( (swiss_roll1, swiss_roll2.dot(rotate_swiss_roll) + 30 ) )
myplot(X)
```

```python
X = np.concatenate( (swiss_roll1, swiss_roll2.dot(rotate_swiss_roll) ))
myplot(X)
```
